```python
import torch
import torch.nn as nn
import torch.optim as optim

# 假设模型和损失函数
model = nn.Linear(10, 1)  # 一个简单的线性模型
loss_fn = nn.MSELoss()    # 损失函数
data = torch.randn(100, 10)  # 输入数据
target = torch.randn(100, 1) # 目标数据

# 梯度下降法
optimizer_gd = optim.SGD(model.parameters(), lr=0.01)

# 随机梯度下降（SGD）
optimizer_sgd = optim.SGD(model.parameters(), lr=0.01)

# 带动量的随机梯度下降（SGDM）
optimizer_sgdm = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

# Adagrad
optimizer_adagrad = optim.Adagrad(model.parameters(), lr=0.01)

# RMSProp
optimizer_rmsprop = optim.RMSprop(model.parameters(), lr=0.01, alpha=0.99)

# Adam
optimizer_adam = optim.Adam(model.parameters(), lr=0.01, betas=(0.9, 0.999), eps=1e-8)

# 示例：训练循环
for epoch in range(10):  # 假设 10 个 epoch
    for x, y in zip(data, target):
        optimizer_adam.zero_grad()  # 清除梯度(可选用其他优化器)
        output = model(x)           # 前向传播
        loss = loss_fn(output, y)   # 计算损失
        loss.backward()             # 反向传播
        optimizer_adam.step()       # 更新参数(可选用其他优化器)
```

```python
import torch
from torch import nn
import torch.nn.functional as F
class LeNet5(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 6, 5, padding=2)
        self.pool1 = nn.MaxPool2d((2, 2))
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.pool2 = nn.MaxPool2d((2, 2))
        self.fc1 = nn.Linear(16*5*5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)
        
        def forward(self, x):
            x = F.relu(self.conv1(x))
            x = self.pool1(x)
            x = F.relu(self.conv2(x))
            x = self.pool2(x)
            x = x.view(len(x), -1)
            x = F.relu(self.fc1(x))
            x = F.relu(self.fc2(x))
            x = self.fc3(x)
            return x
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.datasets import Planetoid

# 定义 GCN 模型
class GCN(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)  # 第一层 GCN 卷积
        self.conv2 = GCNConv(hidden_channels, out_channels)  # 第二层 GCN 卷积

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)  # 第一层卷积
        x = F.relu(x)  # 激活函数
        x = self.conv2(x, edge_index)  # 第二层卷积
        return F.log_softmax(x, dim=1)  # 用于分类任务的 log_softmax 输出

# 加载数据集
dataset = Planetoid(root='/tmp/Cora', name='Cora')  # Planetoid 数据集 (Cora)
data = dataset[0]  # 取出图数据对象

# 初始化模型
in_channels = dataset.num_node_features  # 输入特征维度
hidden_channels = 16  # 隐藏层维度
out_channels = dataset.num_classes  # 输出类别数量
model = GCN(in_channels, hidden_channels, out_channels)

# 定义优化器
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)

# 训练函数
def train():
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)  # 前向传播
    loss = F.nll_loss(out[data.train_mask], data.y[data.train_mask])  # 计算损失
    loss.backward()
    optimizer.step()
    return loss.item()

# 测试函数
def test():
    model.eval()
    out = model(data.x, data.edge_index)
    pred = out.argmax(dim=1)  # 预测类别
    correct = pred[data.test_mask].eq(data.y[data.test_mask]).sum().item()
    acc = correct / data.test_mask.sum().item()
    return acc

# 训练与测试
for epoch in range(200):  # 训练 200 个 epoch
    loss = train()
    if epoch % 10 == 0:  # 每 10 个 epoch 打印一次结果
        acc = test()
        print(f"Epoch: {epoch}, Loss: {loss:.4f}, Test Accuracy: {acc:.4f}")
```
