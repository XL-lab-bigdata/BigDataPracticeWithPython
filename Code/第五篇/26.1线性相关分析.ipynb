```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, kendalltau, pearsonr

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimSun']
plt.rcParams['axes.unicode_minus'] = False

# 1. 读取数据集
data = pd.read_csv('bmi.csv')
data = data.dropna()  # 缺失值处理
data.drop(['Age'], axis=1, inplace=True)  # 删除不需要的列
data.rename(columns={'Height': '身高', 'Weight': '体重', 'Bmi': '体质指数'}, inplace=True)

# 确保数据中只有数值类型的列
numerical_data = data.select_dtypes(include=[np.number])

# 2. 计算相关系数矩阵
corr_methods = {'Pearson': pearsonr, 'Spearman': spearmanr, 'Kendall': kendalltau}
n_features = numerical_data.shape[1]
corr_matrices = {name: np.zeros((n_features, n_features)) for name in corr_methods}

# 循环计算相关系数
for i in range(n_features):
    for j in range(n_features):
        for method, func in corr_methods.items():
            corr_matrices[method][i, j] = func(numerical_data.iloc[:, i], numerical_data.iloc[:, j])[0]

# 转换为 DataFrame
columns = numerical_data.columns
corr_matrices = {name: pd.DataFrame(matrix, columns=columns, index=columns) for name, matrix in corr_matrices.items()}

# 3. 可视化展示相关系数热图
plt.figure(figsize=(12, 4))
for i, (name, corr_matrix) in enumerate(corr_matrices.items()):
    plt.subplot(1, 3, i+1)
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5, cbar=False)
    plt.title(f'{name} 相关系数热图')
plt.tight_layout()
plt.show()

# 4. 简化成对关系图绘制，使用 pairplot
sns.set_palette(['#7C9895', '#C9DCC4', '#DAA87C', '#F4EFAD'])
g = sns.pairplot(numerical_data, diag_kind='hist', markers='o')
g.fig.suptitle('成对关系图', y=1.02)
plt.show()
```
