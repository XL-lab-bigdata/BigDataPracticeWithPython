```python
import numpy as np
import pandas as pd
import xlrd
from mlxtend.frequent_patterns import apriori  # 用于生成频繁项集
from mlxtend.frequent_patterns import association_rules  # 用于生成强关联规则

# 1.导入数据并转换为0-1矩阵
data = pd.read_csv('GoodsOrder.csv', header=0, encoding='gbk')  # 读取CSV文件
order = list(set(data["id"]))  # 获取订单编号列表
goods = list(set(data["Goods"]))  # 获取商品列表

# 2.初始化0-1矩阵
aset = pd.DataFrame(np.zeros((len(order), len(goods))), columns=goods)
for line in data.itertuples():
    order_id = getattr(line, "id")  
    good = getattr(line, "Goods") 
    aset.loc[order_id-1, good] = 1  # 将对应位置设置为1

# 3.调用函数，生成频繁项集与关联规则
if __name__ == '__main__':
    frequent_itemsets = apriori(aset, min_support=0.02, use_colnames=True)  # 设置最小支持度为2%
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.35)  # 设置最小置信度为35%

# 4.结果展示
print(pd.DataFrame(frequent_itemsets)) 
print(pd.DataFrame(rules))
```
