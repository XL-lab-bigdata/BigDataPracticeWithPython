```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# 1.读取数据集
data = pd.read_excel('IT income.xlsx')
X1 = data[['工龄']]
X2 = data['工龄']
Y = data['薪水']

# 2.二次曲线拟合数据
poly = PolynomialFeatures(degree=2)
X11 = poly.fit_transform(X1)
model = LinearRegression()
model.fit(X11, Y)
r2 = r2_score(Y, model.predict(X11))

# 3.幂函数拟合数据
def power_func(x, a, b):
    return a * np.power(x, b)
popt, _ = curve_fit(power_func, X2, Y)
r_squared = r2_score(Y, power_func(X2, *popt))
print("二次曲线 R2:", r2)
print("幂函数 R2:", r_squared)

# 4.绘制图像
fig, axs = plt.subplots(1, 2)
axs[0].plot(X1, model.predict(X11), color='red')
axs[0].scatter(X1, Y)
axs[1].plot(X2, power_func(X2, *popt), 'r-')
axs[1].scatter(X2, Y)
plt.show()
```
