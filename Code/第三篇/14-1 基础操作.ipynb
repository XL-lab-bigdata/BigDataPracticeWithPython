## 14.3.3 通过collection实现文档增删改查

```python
# 选择一个集合
db = client['test_database']
collection = db['example_collection']
# 插入单个文档
collection.insert_one({'name': '胡涛', 'age': 25})
# 插入多个文档
collection.insert_many([
    {'name': '朗晟', 'age': 30},
    {'name': '马迎宾', 'age': 35}
])
```

```python
# 查询单个文档
person = collection.find_one({'name': '胡涛'})
print('find_one结果：',person)

# 查询所有文档
print('find结果：')
for person in collection.find():
    print(person)
```

```python
# 更新单个文档
collection.update_one({"name": "胡涛"}, {"$set": {"age": 30}})
print('胡涛的年龄更新为：', collection.find_one({'name': '胡涛'})['age'])
# 更新多个文档
collection.update_many({"age": 30}, {"$set": {"age": 31}})
for person in collection.find():
    print(person['name']+'的年龄为',person['age'])
```

```python
# 删除单个文档
collection.delete_one({'name': '马迎宾'})
# 删除多个文档
collection.delete_many({'age': {'$gt': 30}})
```

```python
from datetime import datetime, timedelta
import random
# 连接到MongoDB  
client = MongoClient("localhost", 27017)
db = client['sales_db']
sales_collection = db['sales']

# 随机生成10个样例数据  
item_names = ['apple', 'orange', 'banana', 'grape', 'mango']
sample_sales_data = []

# 当前时间  
now = datetime.now()

for _ in range(10):
    item = random.choice(item_names)
    price = round(random.uniform(1, 5), 2)
    quantity = random.randint(1, 30)
    # 随机生成一个过去10天内的时间  
    time = now - timedelta(days=random.randint(0, 10))
    sample_sales_data.append({
        'item': item,
        'price': price,
        'quantity': quantity,
        'time': time
    })

# 清除之前的文档  
try:
    sales_collection.delete_many({})
finally:
    sales_collection.insert_many(sample_sales_data)
```

```python
for record in sales_collection.find():
    print(record)
```

```python
# 定义时间范围和其它过滤条件（例如过去5天内的销售）  
start_date = now - timedelta(days=5)
end_date = now
# 聚合查询计算每种商品的总销售额  
pipeline = [
    {
        '$match': {'time': {
                '$gte': start_date,
                '$lt': end_date
            }
        }
    },
    {
        '$group': {
            '_id': '$item',
            'total_sales': {
                '$sum': {'$multiply': ['$price', '$quantity']}
            }
        }
    },
    {
        '$project': {
            'item': '$_id',
            'total_sales': 1,
            '_id': 0
        }
    }
]
result = list(sales_collection.aggregate(pipeline))
# 输出聚合结果  
for item in result:
    print(f"Item: {item['item']}, Total Sales: {item['total_sales']:.2f}")
```
