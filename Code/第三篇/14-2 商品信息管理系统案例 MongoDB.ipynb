## 14.4.6更新库存和价格

```python
# 更新金枕榴莲的价格
stock_collection.update_one({"fruit_id": 1}, {"$set": {"price": 16.0}})

# 减少金枕榴莲的库存数量
stock_collection.update_one({"fruit_id": 1}, {"$inc": {"stock": -10}})

# 如果某种水果的销量超过100件，将价格增加10%
high_sales_fruits = record_collection.aggregate([
    {"$unwind": "$items"},
    {"$group": {"_id": "$items.fruit", "total_sold": {"$sum": "$items.quantity"}}},
    {"$match": {"total_sold": {"$gt": 100}}}
])

for fruit in high_sales_fruits:
    stock_collection.update_many({"name": fruit["_id"]}, {"$mul": {"price": 1.1}})

# 批量更新，降低所有泰国水果的价格5%
stock_collection.update_many({"origin": "泰国"}, {"$mul": {"price": 0.95}})

# 根据每日销售记录动态调整库存
daily_sales = record_collection.aggregate([
    {"$unwind": "$items"},
    {"$group": {"_id": "$items.fruit", "total_sold": {"$sum": "$items.quantity"}}}
])

for sale in daily_sales:
    # 动态补充库存，补充数量为当天销售数量的50%
    stock_collection.update_many({"name": sale["_id"]}, {"$inc": {"stock": sale["total_sold"] * 0.5}})
```

```python
from datetime import datetime

# 查询某种水果在特定时间范围内的销售数据
def calculate_sales_in_time_range(fruit_name, start_date, end_date):
    pipeline = [
        {"$unwind": "$items"},
        {"$match": {
            "items.fruit": fruit_name,
            "purchase_time": {"$gte": start_date, "$lte": end_date}
        }},
        {
            "$group": {
                "_id": "$items.fruit",
                "total_sales": {"$sum": "$items.total_price"},
                "total_quantity": {"$sum": "$items.quantity"}
            }
        }
    ]
    
    result = list(record_collection.aggregate(pipeline))
    if result:
        return result[0]
    else:
        return {"total_sales": 0, "total_quantity": 0}

# 示例：计算智利车厘子在2024年7月内的销售额和数量
start_date = datetime(2024, 7, 1)
end_date = datetime(2024, 7, 31)
sales_info = calculate_sales_in_time_range("智利车厘子", start_date, end_date)

print(f"销售额：￥{sales_info['total_sales']}, 销售数量：{sales_info['total_quantity']}")
```
