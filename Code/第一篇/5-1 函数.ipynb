## 5.2示例    

```python
def sum_squares (x, y):
	z = x**2+y**2
	return z
# 调用函数
z = sum_squares(3, 4)
print(z)
```

```python
sum_squares(3, 4)
sum_squares(x=3, y=4)
sum_squares(y=4, x=3)
```

## 5.3可变参数函数

```python
numbers = [1, 2, 3]
new_numbers = [*numbers, 4, 5, 6]
print(new_numbers) 
```

```python
defaults = {"color": "red", "size": "medium"}
user_preferences = {"size": "large", "theme": "dark"}
result = {**defaults, **user_preferences}
print(result)
```

#### (1)对可迭代对象进行拆分

```python
print("a", "b", "c")
print(*("a", "b", "c"))             # 元组
print(*["a", "b", "c"])             # 列表
print(*{"a": 1, "b": 2, "c": 3})    # 注意：对字典拆解时只拆解key
print(*"abc")
```

```python
a = *range(3), #逗号不能省略，否则会报错
b = *range(3), 3
c = [*range(3)]
d = {*range(3)}
e = {*{'y': 1}}     # 只对key进行拆解
f = {**{'y': 1}} 
print(a,b,c,d,e,f)
```

#### (2)可变变量的赋值

```python
a, *b, c = {1, 2, 3, 4, 5, 6}
print(a,b,c)
```

#### (3)函数的可选参数标志

```python
def function(a,*args,**kwargs):
    print(a)
    print(args)
    print(kwargs)
# 按照定义顺序首先接收固定参数，然后接收位置参数，最后接收关键词参数
function(1,2,3,c=4)
```

```python
def sum_function(*args):
	print(sum(args))

# 对于这个函数，不管传入多少个参数都是可以的
sum_function(1) # 输出: 1
sum_function(1, 2) # 输出: 3
sum_function(1, 2, 3) # 输出: 6
```

```python
def self_defined_function(*args, **kwargs):
	if "denominator" not in kwargs:
		print(sum(args))
	else:
		# 注意：**会将关键词参数解包为字典赋值到kwargs，故kwargs为字典
		denominator = kwargs["denominator"]
		print(sum(args) / denominator)

# 没有denominator时，打印几个数的和
self_defined_function(1, 2, 3) # 输出: 6
# 有denominator时，先求几个数的和，再除以denominator
self_defined_function(1, 2, 3, denominator=3) # 输出: 2.0
```

## 5.4lambda函数

```python
def sum(num1 , num2):
	print(num1 + num2)
sum(1, 2)
```

```python
sum = lambda num1 , num2 : print(num1 + num2)
sum(1, 2) # 输出: 3
```

```python
add=lambda x, y: x + y # lambda函数赋值给变量add
add(1, 2) # 通过变量add间接调用lambda函数
```

```python
import time
# 定义输出为None的lambda函数替换time.sleep函数
time.sleep = lambda x : None
time.sleep(3)	
#由于lambda输出为None，程序不会休眠3秒钟
```

```python
y = map(lambda x: x ** 2, [1, 2, 3, 4, 5])
for i in y:
	print(i)
```
