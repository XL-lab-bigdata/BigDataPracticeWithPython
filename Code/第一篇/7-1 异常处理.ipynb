## 7.2异常捕获

#### (1)try…except方法

```python
try:
	print('开始程序')
	print(1/0)
except ZeroDivisionError as e:
	print(e)
```

#### (2)try…except...else方法

```python
for i in [-1,0,5]:
	try:
		print(1/i)
	except ZeroDivisionError as e:
		print(e)
	else:
		print('程序正常')	
```

#### (3)try…finally方法

```python
try:
	for i in [2, 1, 0]:
		print(60/i)
except ValueError as v:
	print(v)
except TypeError  as t:
	print(t)
except ZeroDivisionError  as z:
	print(z)
except Exception as e:
	print("捕捉有异常：",  e)
finally: #finally的代码是肯定执行的，不管是否有异常, 但是finally语块是可选的。
	print("必须执行")
```

## 7.3自定义异常类

#### (1)调用Exception类的__init__方法

```python
class AccountError(Exception):
    # AccountError的__init__方法，balance为账户余额，amount为取款金额
    def __init__(self, balance, amount):
        # 父类Exception类的__init__方法，用于设置异常信息
        super().__init__("账户没有{}元".format(amount))
        self.amount = amount
        self.balance = balance
     # 当取出金额大于账户余额时抛出异常
     def overage(self):
         if self.amount > self.balance:
             raise AccountError(self.balance,self.amount)
cs = AccountError(25,50)
cs.overage()
```

#### (2)使用__str__方法

```python
import  datetime
class  ArgumentError(Exception):
    def  __init__(self,  *args,  **kwargs):
            self.args  =  args
            self.kwargs  =  kwargs
    #  若抛出异常则以字符串形式返回输入的参数
    def  __str__(self):
            return  "%s"  %  repr(self.args)  +  repr(self.kwargs)

def  division(a,  b):
    if  b  ==  0:
        raise  ArgumentError("0不能做除数")
    if  not  isinstance(a,  (int,  float))  or  not  isinstance(b,  (int,  float)):
        raise  ArgumentError("参数必须是数值型",  time=datetime.datetime.now().strftime('%F  %H:%M:%S'))
    print(a/b)
division(1,'test')
```
