### 4.4.1 for循环

```python
ESDATE = [2017, 2017, 2018, 2019, 2021, 2024]
# 初始化目标统计值
Num = 0
# 使用 for 循环遍历列表 ESDATE
for date in ESDATE:
    if 2018 <= date <= 2024:
        Num += 1    # 日期符合条件，那么统计目标 Num 增加 1
# 循环结束后，输出统计值      
print(f'2018-2024年，共有{Num}个家庭农场成立。')
```

### 4.4.2 while循环

```python
#使用while判断结构求1~n的和
n=int(input('请输入需要计算的数：'))
sum=0
index=1
while index<=n:
    sum+=index
    index+=1
print("1到%d之和为:%d"%(n,sum))
```

#### （2）自定义迭代器

```python
iterator = iter(courses)
try:
	while True:
		course = next(iterator)
		print(course)
except StopIteration:
	pass
```

```python
class NestedDictIterator:  
    def __init__(self, nested_dict):  
        # 初始化一个栈来存储待遍历的字典和对应的键  
        self.stack = [(nested_dict, None)]  
  
    def __iter__(self):  
        # 返回迭代器对象本身  
        return self  
  
    def __next__(self):  
        # 当栈不为空时，继续遍历  
        while self.stack:  
            # 弹出栈顶元素  
            current_dict, parent_key = self.stack.pop()  
            # 如果当前元素是字典，则进一步遍历其键值对  
            if isinstance(current_dict, dict):  
                for key in current_dict:  
                    # 如果当前键对应的值是字典，则将其压入栈中待后续遍历  
                    if isinstance(current_dict[key], dict):  
                        if parent_key is not None:
                            self.stack.append((current_dict[key], f"{parent_key}.{key}"))  
                        else:
                            self.stack.append((current_dict[key], key))  
                    else:  
                        # 如果当前键对应的值不是字典，则构造完整的键名并返回键值对  
                        if parent_key is not None:  
                            parent_key = f"{parent_key}.{key}"  
                            full_key = parent_key
                        else:  
                            full_key = key  
                        return (full_key, current_dict[key])  
            else:  
                # 如果栈顶元素不是字典，说明遍历结束  
                raise StopIteration  
        # 如果栈为空，说明所有嵌套字典都已遍历完毕  
        raise StopIteration  
  
# 使用示例  
nested_dict = {'a': {'b': {'c': 1}, 'd': 3}, 'e': 4}   
  
# 创建迭代器对象并开始遍历  
for key, value in NestedDictIterator(nested_dict):  
    print(f"{key}: {value}")
```

#### （4）生成器

```python
def fib(max):
	n, a, b = 0, 0, 1
	while n < max:
		yield b
		a, b = b, a + b
		n = n + 1
	return 'done'

nums = fib(10)
iter(nums)
```

```python
next(nums) # 输出: 1
```

### 4.4.4 break和continue语句

```python
for i in range(5):
	if (i==3):
		break
	print(i,end = " ")
```

```python
for i in range(5):
	if (i==3):
		continue
	print(i,end = " ")
```
